```python
remove_list = ["MSSubClass", "LotFrontage", "LotArea","MasVnrArea","BsmtFinSF1","TotalBsmtSF","1stFlrSF",
               "GrLivArea","WoodDeckSF","OpenPorchSF"]
def filter_IQR2(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    print(f"====== {column_name} ======")
    print(f"First Quartile Q1: {Q1}")
    print(f"Third Quartile Q3: {Q3}")
    print(f"The inter-quartile range: {IQR} \n")
    
    print(f"length of data before filtering is {len(df)}")
    lower_range = Q1 - 1.5 * IQR
    upper_range = Q3 + 1.5 * IQR
    
    not_outlier_list = [val for val in df[col] if ((val >= lower_range) & (val <= upper_range))]
    filtered = df.loc[df[col].isin(not_outlier_list)]
    print(f"length of data after filtering is {len(filtered)}\n")

    return filtered.index.tolist()

index_set = set(df.index.tolist())

for col in remove_list:
    index_set = set(filter_IQR2(df, col)).intersection(index_set)
#     print(index_set)

df=df.loc[set(index_set)]
```

```python
import category_encoders as ce
df3 = df2.copy()
enc = ce.OrdinalEncoder(cols=df3.select_dtypes(include=object).columns.tolist())


df3 = enc.fit_transform(df3)
df3
```

```python
import sklearn 

Y = df3.filter(["SalePrice"], axis = 1)
Y.shape

X = df3.drop("SalePrice", 1)
X.shape
  
```

```python
from sklearn.model_selection import train_test_split
train_features, validation_features, train_output, validation_output = train_test_split(X,Y,test_size= 0.33, random_state=42)
```

```python
print(train_features.shape)
print(validation_features.shape)
print(train_output.shape)
print(validation_output.shape)
```

```python
from sklearn.linear_model import LinearRegression
# from sklearn import mean_absolute_error, mean_squared_error

# instantiate the model (using the default parameters)
reg = LinearRegression()

reg.fit(train_features, train_output)

y_pred = reg.predict(train_features)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python

print("MSE=",mean_squared_error(train_output, y_pred))
print("MAE=",mean_absolute_error(train_output, y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_transformed = poly.fit_transform(train_features)

polyn = LinearRegression()
polyn.fit(X_transformed,train_output)
pred = polyn.predict(X_transformed)


print("MSE=",mean_squared_error(train_output, pred))
print("MAE=",mean_absolute_error(train_output, pred))
```

```python
# Visualization
X_sorted, Y_pred = zip(*sorted(zip(train_features, pred), key=lambda x: x[0]))

plt.plot(train_features, train_output, 'ro', ms=10, mec='k')

plt.plot(X_sorted, Y_pred, 'b')

plt.title('Polynomial Regression')

plt.xlabel('X')

plt.ylabel('Y')

plt.show()
```
